```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.flipkart.com/search?q=mobile+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
r=requests.get(url)
soup=BeautifulSoup(r.text,"lxml")
print(soup)

Box= soup.find("div",class_= "_1YokD2 _3Mn1Gg")
```

```python
Product_name=[]
price =[]
description=[]
reviews=[]
```

```python
name= Box.find_all("div",class_= "_4rR01T")
for i in name:
    nam=i.text
    Product_name.append(nam)
print(len(Product_name))
print(Product_name)
```

```python
prices = Box.find_all("div",class_="_30jeq3 _1_WHN1")
for i in prices:
    p=i.text
    price.append(p)
print(len(price))
print(price)
```

```python
rating = Box.find_all("div",class_="_3LWZlK")
for i in rating:
    r=i.text
    reviews.append(r)
print(len(reviews))
print(reviews)
```

```python
descript = Box.find_all("ul",class_="_1xgFaf")
for i in descript:
    d=i.text
    description.append(d)
print(len(description))
print(description)
```

```python
df=pd.DataFrame({"Product_Name":Product_name,
"Price":price,
"Description":description,
"Reviews":reviews})
```
