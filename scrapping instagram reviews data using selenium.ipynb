```python
import pandas as pd
import numpy as np
import selenium as se
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
```

```python
website="https://play.google.com/store/apps/?hl=en_IN&gl=US&pli=1"
```

```python
driver=webdriver.Chrome()
driver.get(website)
```

```python
search= driver.find_element(By.CLASS_NAME,"oc9n0c")
search_button = search.find_element(By.TAG_NAME,"button")
search_button.click()
```

```python
app=input("Enter the app name = ")
input = driver.find_element(By.TAG_NAME,"input")
input.send_keys(app)
```
